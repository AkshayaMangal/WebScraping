```python
import requests
from bs4 import BeautifulSoup
import pandas
import csv
```

```python
pages=list(range(1,29))
description=[]
core=[]
RAM=[]
OS=[]
Storage=[]
Display=[]
Warranty=[]
price=[]
review=[]
for page in pages:
    r=requests.get("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(page)).text
    soup=BeautifulSoup(r,"html.parser") 
    
    desc=soup.find_all("div",class_="_4rR01T")
    for i in range(len(desc)):
        description.append(desc[i].text)
    properties=soup.find_all("li",class_="rgWa7D")
    for i in range(0,len(properties)):
        prop_exp=properties[i].text
        if("Core" in prop_exp):
            core.append(prop_exp)
        elif("DDR4" in prop_exp or "RAM" in prop_exp):
            RAM.append(prop_exp)
        elif("Operating System" in prop_exp):
            OS.append(prop_exp)
        elif("SSD" in prop_exp or"HDD" in prop_exp):
            Storage.append(prop_exp)
        elif("Display" in prop_exp):
            Display.append(prop_exp)
        elif("Warranty"in prop_exp):
            Warranty.append(prop_exp)
    cost=soup.find_all("div",class_="_30jeq3 _1_WHN1")
    for i in range(0,len(cost)):
        price.append(cost[i].text)
    rv=soup.find_all("div",class_="_3LWZlK")
    for i in range(0,len(rv)):
        review.append(rv[i].text)
    
```

```python
print(len(core))
print(len(RAM))
print(len(OS))
print(len(Storage))
print(len(Display))
print(len(Warranty))
print(len(price))
print(len(review))
```

```python
df={'Description':description[630],'Core':core[630],'RAM':RAM[630],'Operating System':OS[slice(630)],'Storage':Storage[slice(630)],'Display':Display[slice(630)],'Warranty':Warranty[slice(630)],'Price':price[slice(630)],'Review':review[slice(630)]}
```

```python
dataset=pandas.DataFrame(data=df)
```
